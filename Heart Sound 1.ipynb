```python
import numpy as np 
import matplotlib.pyplot as plt 
import scipy.io.wavfile as wavfile
import os
i=int(0)
import librosa
import librosa.display
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
hop_length = 64     
n_fft = 1024
for audiofile in os.listdir('Just_murmur/'):
    signal,sr=librosa.load('Just_murmur/'+audiofile,sr=2000)
    mel_signal = librosa.feature.melspectrogram(y=signal, sr=sr, hop_length=hop_length, n_fft=n_fft)
    spectrogram = np.abs(mel_signal)
    power_to_db = librosa.power_to_db(spectrogram, ref=np.max)
    plt.figure(figsize=(8, 7))
    window_size = 1024
    window = np.hanning(window_size)
    stft  = librosa.core.spectrum.stft(signal, n_fft=window_size, hop_length=64, window=window)
    out = 2 * np.abs(stft) / np.sum(window)
    fig = plt.Figure()
    canvas = FigureCanvas(fig)
    ax = fig.add_subplot(111)
    librosa.display.specshow(power_to_db, sr=sr, x_axis='time', y_axis='mel', cmap='magma',hop_length=hop_length)
    s='C:\\Users\\91790\\Desktop\\Heart_Sound\\classification-of-heart-sound-recordings-the-physionet-computing-in-cardiology-challenge-2016-1.0.0\\training\\Training_A\\Abnormal\\'
    s=s+str(i)
    fig.savefig(s+".png")
    i=i+1
    plt.show()
    
```
